```python
import io
import pandas as pd
import numpy as np
percipitation = pd.read_csv(io.BytesIO(uploaded['Daily Precipitation.csv']))
soil = pd.read_csv(io.BytesIO(uploaded['Daily Soil Mositure.csv']))
temperature = pd.read_csv(io.BytesIO(uploaded['Daily Temperature.csv']))
NDVI = pd.read_csv(io.BytesIO(uploaded['Eight Day NDVI.csv']))
production_quantity = pd.read_csv(io.BytesIO(uploaded['Production Quantity.csv']))
```

First I will groupby the percipitation, soil, temperature and NDVI data using year and month. and then I will combine them with the production in respective date for my model building.

```python
percipitation.head()
```

```python
percipitation['year'] = pd.to_datetime(percipitation.start_date).dt.year
percipitation['month'] = pd.to_datetime(percipitation.start_date).dt.month
group1 = percipitation.groupby(['year', 'month']).sum().reset_index()
group1
```

```python
group1.iloc[:,0:3]
```

```python
soil.head()
```

```python
soil['year'] = pd.to_datetime(soil.start_date).dt.year
soil['month'] = pd.to_datetime(soil.start_date).dt.month
group2 = soil.groupby(['year', 'month']).sum().reset_index()
group2
```

```python
group2.iloc[:,0:3]
```

```python
temperature.head()
```

```python
temperature['year'] = pd.to_datetime(temperature.start_date).dt.year
temperature['month'] = pd.to_datetime(temperature.start_date).dt.month
group3 = temperature.groupby(['year', 'month']).mean().reset_index()
group3
```

```python
group3.iloc[:,0:3]
```

```python
NDVI.head()
```

```python
NDVI['year'] = pd.to_datetime(NDVI.start_date).dt.year
NDVI['month'] = pd.to_datetime(NDVI.start_date).dt.month
group4 = NDVI.groupby(['year', 'month']).mean().reset_index()
group4
```

```python
group4.iloc[1:100,0:3]
```

```python
production_quantity.head()
```

```python
production_quantity['year'] = pd.to_datetime(production_quantity.start_date).dt.year
production_quantity['month'] = pd.to_datetime(production_quantity.start_date).dt.month
group5 = production_quantity.groupby(['year', 'month']).sum().reset_index()
group5
```

```python
merge1 = pd.merge(group5, group4, how='left', on=['year', 'month'])
merge1
```

```python
merge2 = pd.merge(group5, group3, how='left', on=['year', 'month'])
merge2
```

```python
merge3 = pd.merge(group5, group2, how='left', on=['year', 'month'])
merge3
```

```python
merge4 = pd.merge(group5, group1, how='left', on=['year', 'month'])
merge4
```

```python
complete = merge4.iloc[:,0:3]
complete['precip'] = merge4.precip
complete['soil'] = merge3.smos
complete['temp'] = merge2.temp
complete['ndvi'] = merge1.ndvi
complete
```

```python
x = complete.iloc[:,3:7]
y = complete.iloc[:,2]
```

After having a complete set of my data, I will use the precip, soil, temp, ndvi to predict the production accordingly and use the model with best score for prediction in the prediction dataset.

```python
import sklearn.linear_model
reg = sklearn.linear_model.LinearRegression().fit(x,y)
reg.score(x,y)
```

```python
re = sklearn.linear_model.Ridge(alpha=.8).fit(x,y)
re.score(x,y)
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(x, y)
clf.score(x,y)
```

It looks that the model is best fit using SVM, so I will choose SVM in the following prediction.
